# Intro

R のよいところ

* フリー，オープンソース
* パッケージが多い
* 最新の解析手法が使える
* データ解析に最適化された言語特性
* コミュニティーが活発
* レポート作成やウェブサイト生成が容易
* IDE
* メタプログラミング
* C，C++，Fortranとの連携

R のよくないところ

* 読みにくいコードコードが多い
* 結果のみにこだわる傾向にある．ソースコードのバージョン管理やテストなどが一般的ではない．
* メタプログラミング
* R本体やパッケージのアップデートによる非互換
* 速度が遅い，メモリを食う

## 対象読者

* R中級者
* 他言語プログラマ．Rがなぜそのような動作をするのか気になる人

## この本から得られるもの

* 基本
* 関数型プログラミング
* メタプログラミング（良い点/悪い点）
* 速いコード，少メモリなコード
* イディオム

## Rのスキルを向上させるメタ・テクニック

* ソースコードを読むこと
* 科学的思考

## 推薦図書

* SICP
* 『コンピュータプログラミングの概念・技法・モデル』
* 『達人プログラマー』

* stackoverflow
* R-help mailing list
